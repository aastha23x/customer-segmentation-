```python
parts = df["Dt_Customer"].str.split("-", n=3, expand=True)
df["day"] = parts[0].astype('int')
df["month"] = parts[1].astype('int')
df["year"] = parts[2].astype('int')
```

```python
for col in df.columns:
    temp = df[col].isnull().sum()
    if temp > 0:
        print(f'Column {col} contains {temp} null values.')
```

```python
df = df.dropna()
print("Total values in the dataset after removing the null values:", len(df))
```

```python
#Label Encoding is used to convert the categorical values 
#into the numerical values so that model can understand it.
```

```python
for col in df.columns:
    if df[col].dtype==object:
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col])
```

```python
scaler = StandardScaler()
data = scaler.fit_transform(df)
```

```python
customer = []
for n_clusters in range(1, 21):
    model = KMeans(init='k-means++',
                   n_clusters=n_clusters,
                   max_iter=500,
                   random_state=22)
    model.fit(df)
    customer.append(model.inertia_)
```

```python
plt.figure(figsize=(10, 5))
sb.lineplot(x=range(1, 21), y=customer)
sb.scatterplot(x=range(1, 21), y=customer)
plt.show()
```

```python
model = KMeans(init='k-means++',
               n_clusters=5,
               max_iter=500,
               random_state=22)
segments = model.fit_predict(df)
```

```python
tsne = TSNE(n_components=2, random_state=42)
tsne_data = tsne.fit_transform(df)
```

```python
df_tsne = pd.DataFrame({'x': tsne_data[:, 0], 'y': tsne_data[:, 1], 'segment': segments})

plt.figure(figsize=(7, 7))
sb.scatterplot(x='x', y='y', hue='segment', data=df_tsne, palette='viridis')
plt.title('t-SNE Visualization of Customer Segments')
plt.show()
```
